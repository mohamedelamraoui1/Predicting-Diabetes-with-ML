# Description du Dataset de Pr√©diction du Diab√®te

Le dataset **Diabetes Prediction** contient des donn√©es m√©dicales et d√©mographiques de patients, ainsi que leur statut diab√©tique (positif ou n√©gatif). Voici une explication simple de chaque colonne :

- **√Çge** : L‚Äô√¢ge du patient, allant de 0 √† 80 ans. Plus on est √¢g√©, plus le risque de diab√®te peut augmenter.
- **Genre** : Le sexe du patient (Homme, Femme, ou Autre). Cela peut influencer la probabilit√© du diab√®te.
- **IMC (BMI)** : Indice de masse corporelle, entre 10.16 et 71.55. Un IMC √©lev√© est souvent li√© au diab√®te.
- **Hypertension** : Indique si le patient a une pression art√©rielle √©lev√©e (0 = non, 1 = oui). C‚Äôest un facteur de risque.
- **Maladie cardiaque** : Indique si le patient a une maladie du c≈ìur (0 = non, 1 = oui). Li√© au diab√®te dans certains cas.
- **Historique de tabagisme** : D√©crit les habitudes de tabac (non-fumeur, ancien fumeur, actuel, etc.). Fumer peut aggraver le risque.
- **Niveau HbA1c** : Mesure le sucre dans le sang sur 2-3 mois (3.5 √† 9%). Un niveau √©lev√© signale un risque de diab√®te.
- **Niveau de glucose sanguin** : Quantit√© de sucre dans le sang √† un moment donn√© (80 √† 300 mg/dL). Valeurs hautes = signe de diab√®te.
- **Diab√®te** : La variable √† pr√©dire (0 = pas de diab√®te, 1 = diab√®te). C‚Äôest l‚Äôobjectif principal.

```python
import pandas as pd

# Charger le dataset
diabetes_dataset = pd.read_csv("diabetes_prediction_dataset.csv")

# Afficher les premi√®res lignes pour un aper√ßu
print("Aper√ßu des 5 premi√®res lignes du dataset :")
print(diabetes_dataset.head())

# Informations g√©n√©rales sur le dataset (types de donn√©es, valeurs non nulles)
print("\nInformations g√©n√©rales sur le dataset :")
print(diabetes_dataset.info())

# Statistiques descriptives pour les colonnes num√©riques
print("\nStatistiques descriptives des colonnes num√©riques :")
print(diabetes_dataset.describe())

# V√©rifier le nombre de valeurs manquantes par colonne
print("\nNombre de valeurs manquantes par colonne :")
print(diabetes_dataset.isnull().sum())


```

**Remarque** : Aucune valeur manquante d√©tect√©e dans le dataset (0 pour toutes les colonnes).

```python
from sklearn.preprocessing import OneHotEncoder

# V√©rifier les valeurs uniques des variables cat√©goriques
print("Valeurs uniques de 'gender' :")
print(diabetes_dataset['gender'].unique())
print("\nValeurs uniques de 'smoking_history' :")
print(diabetes_dataset['smoking_history'].unique())

# S√©parer les colonnes cat√©goriques √† encoder
categorical_columns = ['gender', 'smoking_history']

# Initialiser l'encodeur One-Hot
encoder = OneHotEncoder(sparse_output=False) 

# Appliquer l'encodage sur les colonnes cat√©goriques
encoded_data = encoder.fit_transform(diabetes_dataset[categorical_columns])

# Obtenir les noms des nouvelles colonnes encod√©es
encoded_column_names = encoder.get_feature_names_out(categorical_columns)

# Cr√©er un DataFrame avec les donn√©es encod√©es
encoded_df = pd.DataFrame(encoded_data, columns=encoded_column_names)

# Supprimer les colonnes cat√©goriques originales et ajouter les encod√©es
diabetes_encoded = diabetes_dataset.drop(categorical_columns, axis=1)
diabetes_encoded = pd.concat([diabetes_encoded, encoded_df], axis=1)

# Afficher un aper√ßu du nouveau dataset encod√©
print("\nAper√ßu du dataset apr√®s encodage :")
print(diabetes_encoded.head())

# V√©rifier les colonnes du dataset final
print("\nColonnes du dataset encod√© :")
print(diabetes_encoded.columns)
```

```python
# Dataset avec valeurs aberrantes (non trait√©)
diabetes_encoded_with_outliers = diabetes_encoded.copy()

# Colonnes num√©riques √† analyser
numeric_columns = ['age', 'bmi', 'HbA1c_level', 'blood_glucose_level']

# Seuils r√©alistes bas√©s sur le domaine m√©dical
realistic_bounds = {
    'age': (0, 100),              # √Çge plausible
    'bmi': (10, 60),              # BMI extr√™mes mais possibles
    'HbA1c_level': (3, 12),       # HbA1c m√©dicalement plausible
    'blood_glucose_level': (50, 400)  # Glucose plausible
}

# Fonction pour d√©tecter et cr√©er un dataset sans valeurs aberrantes
def detect_and_remove_outliers(df, column, lower_bound, upper_bound):
    # Compter les aberrants selon les seuils r√©alistes
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)][column]
    print(f"\nColonne '{column}' :")
    print(f" - Limite r√©aliste appliqu√©e : [{lower_bound}, {upper_bound}]")
    print(f" - Nombre de valeurs aberrantes : {len(outliers)}")
    
    # Supprimer les lignes avec valeurs aberrantes
    df_cleaned = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df_cleaned

# Cr√©er le dataset sans valeurs aberrantes
diabetes_encoded_no_outliers = diabetes_encoded.copy()
for col in numeric_columns:
    lower, upper = realistic_bounds[col]
    diabetes_encoded_no_outliers = detect_and_remove_outliers(diabetes_encoded_no_outliers, col, lower, upper)

# Afficher les dimensions des deux datasets
print("\nDimensions du dataset avec valeurs aberrantes :", diabetes_encoded_with_outliers.shape)
print("Dimensions du dataset sans valeurs aberrantes :", diabetes_encoded_no_outliers.shape)

```

**R√©sultat** : Deux datasets cr√©√©s : `diabetes_encoded_with_outliers` (avec aberrants) et `diabetes_encoded_no_outliers` (sans aberrants) pour comparer leur impact sur les mod√®les.

##  Application de KNN + Visualisation + √âvaluation

Nous entra√Ænons un mod√®le KNN avec validation crois√©e (K-Fold, K=5) et utilisons l'ACP pour visualiser les groupes form√©s par les pr√©dictions de KNN sur les deux datasets (avec et sans outliers). Les performances sont √©valu√©es avec accuracy, F1-score, pr√©cision et rappel, suivies d'une matrice de confusion et d'une visualisation PCA.

```python
# Importer les biblioth√®ques n√©cessaires
import time
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score, KFold
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.decomposition import PCA


# D√©finir le mod√®le KNN
knn = KNeighborsClassifier(n_neighbors=5)

# Configurer la validation crois√©e (K-Fold, K=5)
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Fonction pour l'√©chantillonnage
def sample_data(X, y, sample_frac=0.1):
    """Effectue un √©chantillonnage al√©atoire des donn√©es X et y avec la fraction sp√©cifi√©e."""
    X_sample = X.sample(frac=sample_frac, random_state=42)
    y_sample = y[X_sample.index]
    return X_sample, y_sample

# Fonction pour entra√Æner, √©valuer et visualiser avec PCA
def evaluate_knn(X, y, dataset_name, sample_frac=0.1):
    print(f"\n√âvaluation de KNN sur {dataset_name} (√©chantillon {int(sample_frac*100)}%) :")
    
    # √âchantillonnage des donn√©es
    X_sample, y_sample = sample_data(X, y, sample_frac)
    
    # Cross validation pour plusieurs m√©triques avec mesure du temps
    print("Cross Validation :")
    start_time = time.time()  # D√©but du chronom√®tre
    
    metrics = ['accuracy', 'f1', 'precision', 'recall']
    for metric in metrics:
        scores = cross_val_score(knn, X_sample, y_sample, cv=kf, scoring=metric)
        print(f"{metric.capitalize()} :")
        print(f" - Scores : {scores}")
        print(f" - Moyenne : {scores.mean():.3f}")
        print(f" - √âcart-type : {scores.std():.3f}\n")
    # Entra√Æner le mod√®le sur l'√©chantillon
    print("Entra√Ænement et Validation :")
    knn.fit(X_sample, y_sample)
    y_pred = knn.predict(X_sample)
    # Fin du chronom√®tre apr√®s l'entra√Ænement et la pr√©diction
    end_time = time.time()  # Fin du chronom√®tre
    elapsed_time = end_time - start_time
    print(f"Temps de calcul total (Cross Validation + Entra√Ænement/Validation) : {elapsed_time:.2f} secondes\n")
    # Rapport de classification
    print("Rapport de classification :")
    print(classification_report(y_sample, y_pred))

    # Matrice de confusion
    cm = confusion_matrix(y_sample, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title(f"Matrice de confusion pour KNN ({dataset_name})")
    plt.xlabel("Pr√©diction")
    plt.ylabel("V√©rit√©")
    plt.show()

    # Visualisation des groupes avec PCA
    print("Visualisation avec PCA :")
    pca = PCA(n_components=2)  # R√©duire √† 2 dimensions
    X_pca = pca.fit_transform(X_sample)
    
    plt.figure(figsize=(6, 4))
    scatter = sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=y_pred, palette='Set1', s=50)
    plt.title(f"Visualisation des groupes KNN avec PCA ({dataset_name})")
    plt.xlabel("Composante Principale 1")
    plt.ylabel("Composante Principale 2")
    plt.legend(title="Pr√©diction", labels=["Non diab√©tique (0)", "Diab√©tique (1)"])
    plt.show()

    # Afficher la variance expliqu√©e
    print(f"Variance expliqu√©e par les composantes principales ({dataset_name}) :")
    print(pca.explained_variance_ratio_)
    
    return elapsed_time  # Retourner le temps pour une utilisation future

```

```python
# Appliquer sur les deux datasets et r√©cup√©rer les temps
knn_time_with_outliers = evaluate_knn(X_with_outliers_scaled, y_with_outliers, "dataset avec outliers")
knn_time_no_outliers = evaluate_knn(X_no_outliers_scaled, y_no_outliers, "dataset sans outliers")

# Afficher les temps pour confirmation
print(f"\nTemps de calcul total KNN (avec outliers) : {knn_time_with_outliers:.2f} secondes")
print(f"Temps de calcul total KNN (sans outliers) : {knn_time_no_outliers:.2f} secondes")
```

##  Application de SVM + Visualisation + √âvaluation du Mod√®le

Nous entra√Ænons un mod√®le SVM avec validation crois√©e (K-Fold, K=5) sur les deux datasets (avec et sans outliers). SVM est puissant pour les donn√©es lin√©aires ou non lin√©aires (avec kernel), offrant une alternative √† KNN. La visualisation (via PCA) et l‚Äô√©valuation avec F1-score, pr√©cision, rappel et accuracy permettent de juger son efficacit√© face au d√©s√©quilibre des classes (91.5 % de `0` vs 8.5 % de `1`). Cette √©tape suit KNN pour comparer une approche plus complexe et sensible au scaling.

**Justification de l‚Äôordre** : SVM suit KNN car il est plus complexe et d√©pend fortement du scaling, ce qui a d√©j√† √©t√© appliqu√©. Cela permet une comparaison directe avec KNN tout en explorant une m√©thode diff√©rente.

```python
# Importer les biblioth√®ques n√©cessaires
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score, KFold
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.decomposition import PCA


# D√©finir le mod√®le SVM
svm = SVC(kernel='rbf', random_state=42, probability=True)  # Kernel RBF, ajustable

# Configurer la validation crois√©e (K-Fold, K=5)
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Fonction pour l'√©chantillonnage
def sample_data(X, y, sample_frac=0.1):
    """Effectue un √©chantillonnage al√©atoire des donn√©es X et y avec la fraction sp√©cifi√©e."""
    X_sample = X.sample(frac=sample_frac, random_state=42)
    y_sample = y[X_sample.index]
    return X_sample, y_sample

# Fonction pour entra√Æner, √©valuer et visualiser avec PCA
def evaluate_svm(X, y, dataset_name, sample_frac=0.1):
    print(f"\n√âvaluation de SVM sur {dataset_name} (√©chantillon {int(sample_frac*100)}%) :")
    
    # √âchantillonnage des donn√©es
    X_sample, y_sample = sample_data(X, y, sample_frac)
    
    # D√©but du chronom√®tre pour tout le processus
    start_time = time.time()
    
    # Cross validation pour plusieurs m√©triques
    print("Cross Validation :")
    metrics = ['accuracy', 'f1', 'precision', 'recall']
    for metric in metrics:
        scores = cross_val_score(svm, X_sample, y_sample, cv=kf, scoring=metric)
        print(f"{metric.capitalize()} :")
        print(f" - Scores : {scores}")
        print(f" - Moyenne : {scores.mean():.3f}")
        print(f" - √âcart-type : {scores.std():.3f}\n")
    
    # Entra√Ænement et Validation
    print("Entra√Ænement et Validation :")
    svm.fit(X_sample, y_sample)
    y_pred = svm.predict(X_sample)
    
    # Fin du chronom√®tre apr√®s l'entra√Ænement et la pr√©diction
    end_time = time.time()
    elapsed_time = end_time - start_time
    print(f"Temps de calcul total (Cross Validation + Entra√Ænement/Validation) : {elapsed_time:.2f} secondes\n")
    
    # Rapport de classification
    print("Rapport de classification :")
    print(classification_report(y_sample, y_pred))

    # Matrice de confusion
    cm = confusion_matrix(y_sample, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title(f"Matrice de confusion pour SVM ({dataset_name})")
    plt.xlabel("Pr√©diction")
    plt.ylabel("V√©rit√©")
    plt.show()

    # Visualisation des groupes avec PCA
    print("Visualisation avec PCA :")
    pca = PCA(n_components=2)  # R√©duire √† 2 dimensions
    X_pca = pca.fit_transform(X_sample)
    
    plt.figure(figsize=(6, 4))
    scatter = sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=y_pred, palette='Set1', s=50)
    plt.title(f"Visualisation des groupes SVM avec PCA ({dataset_name})")
    plt.xlabel("Composante Principale 1")
    plt.ylabel("Composante Principale 2")
    plt.legend(title="Pr√©diction", labels=["Non diab√©tique (0)", "Diab√©tique (1)"])
    plt.show()

    # Afficher la variance expliqu√©e
    print(f"Variance expliqu√©e par les composantes principales ({dataset_name}) :")
    print(pca.explained_variance_ratio_)
    
    return elapsed_time  # Retourner le temps total
```

```python
# Appliquer sur les deux datasets et r√©cup√©rer les temps
svm_time_with_outliers = evaluate_svm(X_with_outliers_scaled, y_with_outliers, "dataset avec outliers")
svm_time_no_outliers = evaluate_svm(X_no_outliers_scaled, y_no_outliers, "dataset sans outliers")

# Afficher les temps pour confirmation
print(f"\nTemps de calcul total SVM (avec outliers) : {svm_time_with_outliers:.2f} secondes")
print(f"Temps de calcul total SVM (sans outliers) : {svm_time_no_outliers:.2f} secondes")
```

##  Application de DT + RF + Visualisation + √âvaluation du Mod√®le

Nous entra√Ænons un mod√®le d'Arbre de D√©cision (DT) et un mod√®le de For√™t Al√©atoire (RF) avec validation crois√©e (K-Fold, K=5) sur les deux datasets (avec et sans outliers).
```python
# Importer les biblioth√®ques n√©cessaires
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score, KFold
from sklearn.metrics import classification_report, confusion_matrix

# D√©finir les mod√®les
dt = DecisionTreeClassifier(random_state=42)
rf = RandomForestClassifier(n_estimators=100, random_state=42)

# Configurer la validation crois√©e (K-Fold, K=5)
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Fonction pour l'√©chantillonnage
def sample_data(X, y, sample_frac=0.1):
    """Effectue un √©chantillonnage al√©atoire des donn√©es X et y avec la fraction sp√©cifi√©e."""
    X_sample = X.sample(frac=sample_frac, random_state=42)
    y_sample = y[X_sample.index]
    return X_sample, y_sample

# Fonction pour la cross validation
def perform_cross_validation(model, X, y, dataset_name, sample_frac=0.1):
    print(f"\nCross Validation de {model.__class__.__name__} sur {dataset_name} (√©chantillon {int(sample_frac*100)}%) :")
    X_sample, y_sample = sample_data(X, y, sample_frac)
    # D√©but du chronom√®tre
    start_time = time.time()
    metrics = ['accuracy', 'f1', 'precision', 'recall']
    for metric in metrics:
        scores = cross_val_score(model, X_sample, y_sample, cv=kf, scoring=metric)
        print(f"{metric.capitalize()} :")
        print(f" - Scores : {scores}")
        print(f" - Moyenne : {scores.mean():.3f}")
        print(f" - √âcart-type : {scores.std():.3f}\n")
    # Fin du chronom√®tre
    end_time = time.time()
    elapsed_time = end_time - start_time
    print(f"Temps de calcul pour la Cross Validation : {elapsed_time:.2f} secondes\n")
    
    return elapsed_time  # Retourner le temps
    

# Fonction pour la validation (entra√Ænement et √©valuation d√©taill√©e)
def perform_validation(model, X, y, dataset_name, sample_frac=0.1):
    print(f"\nValidation de {model.__class__.__name__} sur {dataset_name} (√©chantillon {int(sample_frac*100)}%) :")
    X_sample, y_sample = sample_data(X, y, sample_frac)
    # D√©but du chronom√®tre
    start_time = time.time()
    
    model.fit(X_sample, y_sample)
    y_pred = model.predict(X_sample)

    # Fin du chronom√®tre
    end_time = time.time()
    elapsed_time = end_time - start_time
    print(f"Temps de calcul pour l'Entra√Ænement et Validation : {elapsed_time:.2f} secondes\n")
    
    print("Rapport de classification :")
    print(classification_report(y_sample, y_pred))
    
    cm = confusion_matrix(y_sample, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title(f"Matrice de confusion pour {model.__class__.__name__} ({dataset_name})")
    plt.xlabel("Pr√©diction")
    plt.ylabel("V√©rit√©")
    plt.show()
    return elapsed_time  # Retourner le temps

# Fonction pour la visualisation
def visualize_groups(model, X, y, dataset_name, sample_frac=0.1):
    print(f"\nVisualisation des groupes {model.__class__.__name__} sur {dataset_name} (√©chantillon {int(sample_frac*100)}%) :")
    X_sample, y_sample = sample_data(X, y, sample_frac)
    model.fit(X_sample, y_sample)
    y_pred = model.predict(X_sample)
    
    pca = PCA(n_components=2)
    X_pca = pca.fit_transform(X_sample)
    
    plt.figure(figsize=(6, 4))
    scatter = sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=y_pred, palette='Set1', s=50)
    plt.title(f"Visualisation des groupes {model.__class__.__name__} avec PCA ({dataset_name})")
    plt.xlabel("Composante Principale 1")
    plt.ylabel("Composante Principale 2")
    plt.legend(title="Pr√©diction", labels=["Non diab√©tique (0)", "Diab√©tique (1)"])
    plt.show()
    
    print(f"Variance expliqu√©e par les composantes principales ({dataset_name}) :")
    print(pca.explained_variance_ratio_)

```
